```python
#Basic import
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd   
import seaborn as sns  

#modelling
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor, CatBoostClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
```

```python
#dataset loading
df=pd.read_csv('Data\StudentsPerformance.csv')
display(df.sample(10))
```

```python
#column rename
df.rename(columns={'race/ethnicity':'race_ethnicity',
                'parental level of education':'parental_education',
                'test preparation course':'test_prep_course',
                'math score':'math_score',
                'reading score':'reading_score',
                'writing score':'writing_score'
                }, inplace=True)

display(df.columns)
```

```python
x=df.drop(columns=['math_score'], axis=1)
y=df['math_score']
```

```python
#for independent data--Feature Transformation
num_feature=x.select_dtypes(exclude='object').columns
cat_feature=x.select_dtypes(include='object').columns

standardscale=StandardScaler()
ohe=OneHotEncoder()


transformer=ColumnTransformer(
    [
        ('OneHotEncoder',ohe,cat_feature),
        ('StandardScaler',standardscale,num_feature),
    ]
)
```

```python
x2=x.copy()
y2=y.copy()
```

```python
x_transformed=transformer.fit_transform(x)
display(x_transformed)
```

```python
x_train, x_test, y_train,y_test=train_test_split(x2,y2, 
                                                test_size=0.25,
                                                random_state=42)

display(x_train,x_test)
display(y_train,y_test)
```

```python
x_train2=x_train.copy()
x_test2=x_test.copy()
y_train2=y_train.copy()
y_test2=y_test.copy()
```

```python
x_train_transformed=transformer.fit_transform(x_train2)
x_test_transformed=transformer.fit_transform(x_test2)

display(x_train_transformed, x_test_transformed)
```

```python
x_train_transformed2=x_train_transformed.copy()
x_test_transformed2=x_test_transformed.copy()
```

```python
#creating a evaluating function

def regression_evaluation(actual, predicted):
    mae=mean_absolute_error(actual, predicted)
    mse=mean_squared_error(actual, predicted)
    rmse=np.sqrt(mse)
    r2=r2_score(actual, predicted)
    return mae,rmse,r2
```

```python
#model

models={
    "LinearRegression":LinearRegression(),
    "Lasso":Lasso(),
    "Ridge":Ridge(),
    "K-Neighbors Regressor":KNeighborsRegressor(),
    "DecisionTree Regressor":DecisionTreeRegressor(),
    "RandomForest Regressor":RandomForestRegressor(),
    "XGBRegressor":XGBRegressor(),
    "AdaBoost Regressor":AdaBoostRegressor(),
    "GradientBoost Regressor":GradientBoostingRegressor(),
    "Catboost Regressor":CatBoostRegressor()
    
}
```

```python
list(models) #generate list of key from the key-value pairs in a dictionary
```

```python
list(models.values())
```

```python
model_list=[]
r2_list=[]
results=[]


for i in range(len(list(models))):
    model=list(models.values())[i]
    model_name=list(models.keys())[i]
    
    #fit the model
    model.fit(x_train_transformed2,y_train2)
    
    #prediction
    y_train_pred=model.predict(x_train_transformed2)
    y_test_pred=model.predict(x_test_transformed2)
    
    #Evaluation
    #Train data
    model_train_mae,model_train_rmse,model_train_r2=regression_evaluation(y_train2,y_train_pred)
    #test data
    model_test_mae,model_test_rmse,model_test_r2=regression_evaluation(y_test2,y_test_pred)
    
    #Store results
    results.append({
        'model_name': model_name,
        'mae_train': model_train_mae,
        'mae_test': model_test_mae,
        'rmse_train': model_train_rmse,
        'rmse_test': model_test_rmse,
        'r2_train': model_train_r2,
        'r2_test': model_test_r2
    })

# Convert to DataFrame
results_df = pd.DataFrame(results)

# Optionally, sort by best R2 on test data
results_df = results_df.sort_values(by='r2_test', ascending=False)
results_df=results_df.reset_index(drop=True)

# Display
display(results_df)
```
